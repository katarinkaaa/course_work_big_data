```python
print(df.info())

print(df.describe())

print(df.isnull().sum())
```

```python
# Целевая переменная 'Churn' - переведем в бинарный вид 1/0
df['Churn'] = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
# Анализ целевой переменной
plt.figure(figsize=(6,4))
sns.countplot(x='Churn', data=df)
plt.title('Распределение целевой переменной Churn')
plt.show()
churn_rate = df['Churn'].value_counts(normalize=True)
print(f"Процент оттока клиентов: {churn_rate[1]:.2%}")
```

```python
# Обнаружена проблема: столбец TotalCharges имеет тип object, хотя должен быть числовым. Есть пустые строки.
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
print(f"Найдено пропусков в TotalCharges после конвертации: {df['TotalCharges'].isnull().sum()}")

# Заполним пропуски в TotalCharges медианным значением
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)
```

```python
# Визуализация распределений
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# tenure
df[df['Churn'] == 0]['tenure'].hist(bins=30, alpha=0.7, label='Не ушел', ax=axes[0,0], color='green')
df[df['Churn'] == 1]['tenure'].hist(bins=30, alpha=0.7, label='Ушел', ax=axes[0,0], color='red')
axes[0,0].set_title('Распределение времени жизни клиента (Tenure)')
axes[0,0].set_xlabel('Месяцы')
axes[0,0].set_ylabel('Количество клиентов')
axes[0,0].legend()

# MonthlyCharges
df[df['Churn'] == 0]['MonthlyCharges'].hist(bins=30, alpha=0.7, label='Не ушел', ax=axes[0,1], color='green')
df[df['Churn'] == 1]['MonthlyCharges'].hist(bins=30, alpha=0.7, label='Ушел', ax=axes[0,1], color='red')
axes[0,1].set_title('Распределение ежемесячных трат (MonthlyCharges)')
axes[0,1].set_xlabel('Стоимость ($)')
axes[0,1].set_ylabel('Количество клиентов')
axes[0,1].legend()

# Contract
contract_churn = pd.crosstab(df['Contract'], df['Churn'], normalize='index')
contract_churn.columns = ['Не ушел', 'Ушел']
contract_churn.plot(kind='bar', ax=axes[1,0], color=['green', 'red'])
axes[1,0].set_title('Влияние типа контракта на отток')
axes[1,0].set_ylabel('Доля клиентов')
axes[1,0].tick_params(axis='x', rotation=45)

# InternetService
internet_churn = pd.crosstab(df['InternetService'], df['Churn'], normalize='index')
internet_churn.columns = ['Не ушел', 'Ушел']
internet_churn.plot(kind='bar', ax=axes[1,1], color=['green', 'red'])
axes[1,1].set_title('Влияние типа интернет-услуги на отток')
axes[1,1].set_ylabel('Доля клиентов')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

# Гипотезы

1.  Клиенты с помесячным контрактом (Month-to-month) уходят значительно чаще, чем клиенты с долгосрочными контрактами (One year, Two year).

2.  Клиенты с оптоволоконным интернетом (Fiber optic) демонстрируют более высокий процент оттока, чем клиенты с DSL или без интернета. Возможная причина: высокая цена и/или нестабильность сервиса. необходимо проанализировать ценовую политику и качество услуги Fiber optic

3.  Время жизни клиента (tenure) является ключевым фактором. Новые клиенты (с низким tenure) и, возможно, клиенты "ветераны" на определенном этапе более склонны к уходу. нужны специальные программы адаптации для новых клиентов и программы поощрения лояльности для долгосрочных

4.  Клиенты, использующие безбумажный биллинг (Paperless Billing), уходят чаще.

5.  Способ оплаты (Payment Method) сильно влияет на отток. Клиенты, оплачивающие электронными чеками, могут уходить чаще.

6. Клиенты с услугой трансляции фильмов демонстрируют более низкий отток


7. Клиенты с техподдержкой имеют значительно более низкий процент оттока
